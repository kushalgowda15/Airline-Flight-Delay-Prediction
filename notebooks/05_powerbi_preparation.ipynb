```python
import pandas as pd
from pathlib import Path
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Determine project root robustly (works in scripts and notebooks)
try:
    ROOT = Path(__file__).resolve().parents[1]
except Exception:
    # __file__ is not defined in notebooks. Walk upwards from cwd to find the repo root
    cwd = Path.cwd()
    search = cwd
    ROOT = None
    while True:
        if (search / 'dataset').exists() and (search / 'models').exists():
            ROOT = search
            break
        if search.parent == search:
            break
        search = search.parent
    if ROOT is None:
        # Last-resort fallback to known absolute path used in this workspace
        ROOT = Path(r'C:\Users\gowda\Downloads\FDS PROJECT\Airline_Flight_Delay_Prediction')

DATA = ROOT / 'dataset' / 'model_predictions.csv'
OUT = ROOT / 'reports' / 'metrics_comparison.csv'

# Load
df = pd.read_csv(DATA)

# Ensure baseline exists (use the same baseline used earlier)
if 'Predicted_Baseline' not in df.columns:
    df['Predicted_Baseline'] = (df['Predicted_Prob'] >= 0.5).astype(int)

# Prepare variables expected by the user's snippet
# y_test: true labels
# For Random Forest: use Predicted_RF and Predicted_Prob
# For Logistic Regression: use Predicted_Baseline (thresholded) and Predicted_Prob as a probability proxy

y_test = df['Actual'].astype(int)
y_pred_rf = df['Predicted_RF'].astype(int)
y_prob_rf = df['Predicted_Prob'].astype(float)

y_pred_lr = df['Predicted_Baseline'].astype(int)
y_prob_lr = df['Predicted_Prob'].astype(float)

# Build metrics DataFrame (user's code)
metrics = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-score', 'ROC-AUC'],
    'Logistic Regression': [
        accuracy_score(y_test, y_pred_lr),
        precision_score(y_test, y_pred_lr, zero_division=0),
        recall_score(y_test, y_pred_lr, zero_division=0),
        f1_score(y_test, y_pred_lr, zero_division=0),
        roc_auc_score(y_test, y_prob_lr)
    ],
    'Random Forest': [
        accuracy_score(y_test, y_pred_rf),
        precision_score(y_test, y_pred_rf, zero_division=0),
        recall_score(y_test, y_pred_rf, zero_division=0),
        f1_score(y_test, y_pred_rf, zero_division=0),
        roc_auc_score(y_test, y_prob_rf)
    ]
})

# Print and save
print(metrics.to_string(index=False))

OUT.parent.mkdir(parents=True, exist_ok=True)
metrics.to_csv(OUT, index=False)
print(f"\nSaved metrics summary to: {OUT.resolve()}")
```
