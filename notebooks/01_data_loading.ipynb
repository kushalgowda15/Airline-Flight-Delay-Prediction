# 01 - Data Loading
Load the dataset and inspect the first rows.

```python
# ---------------------------------------------------------
# Notebook: 01_data_loading.ipynb
# Purpose: Load and inspect the raw flight delay dataset
# Dataset: Airline_Delay_Cause.csv
# ---------------------------------------------------------

# ===============================
# 1️ Imports and basic setup
# ===============================
import pandas as pd
import numpy as np
import os
import time

# Display options
pd.set_option('display.max_columns', 200)
pd.set_option('display.float_format', '{:,.2f}'.format)

# ===============================
# 2️ Define dataset path
# ===============================
DATA_PATH = r"C:\Users\gowda\Downloads\FDS PROJECT\Airline_Flight_Delay_Prediction\dataset\Airline_Delay_Cause.csv"
SNAPSHOT_FILE = r"C:\Users\gowda\Downloads\FDS PROJECT\Airline_Flight_Delay_Prediction\dataset\flights_initial_snapshot.csv"

# ===============================
# 3️ Load dataset
# ===============================
print("="*25, "Loading Dataset", "="*25)

if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f" File not found at: {DATA_PATH}")

start = time.time()
df = pd.read_csv(DATA_PATH, low_memory=False)
print(f" Dataset successfully loaded in {time.time() - start:.2f}s → {df.shape[0]:,} rows × {df.shape[1]} columns")

# ===============================
# 4️ Quick overview
# ===============================
print("\n Column Names:")
print(df.columns.tolist())

print("\n First 5 Rows:")
display(df.head())

print("\n Dataset Info:")
df.info()

# ===============================
# 5️ Basic Sanity Summary
# ===============================
print("\n" + "="*25, "Sanity Summary", "="*25)

if 'year' in df.columns:
    print(f" Years in data: {sorted(df['year'].unique())}")
if 'month' in df.columns:
    print(f" Months range: {df['month'].min()} → {df['month'].max()}")
if 'carrier' in df.columns:
    print(f" Unique carriers: {df['carrier'].nunique()}")
    print(" Sample carriers:", df['carrier'].unique()[:10])
if 'airport' in df.columns:
    print(f" Unique airports: {df['airport'].nunique()}")
    print(" Sample airports:", df['airport'].unique()[:10])
if 'arr_flights' in df.columns:
    print(f" Total recorded arrival flights: {df['arr_flights'].sum():,}")
if 'arr_delay' in df.columns:
    print(f" Average arrival delay: {df['arr_delay'].mean():.2f} minutes")
if 'arr_del15' in df.columns:
    print(f" Delayed (15+ min) flights: {df['arr_del15'].sum():,}")

# ===============================
# 6️ Missing value summary
# ===============================
print("\n" + "="*25, "Missing Value Summary", "="*25)
missing_summary = df.isna().sum().reset_index().rename(columns={'index': 'column', 0: 'missing_count'})
missing_summary['missing_%'] = (missing_summary['missing_count'] / len(df) * 100).round(2)
missing_summary = missing_summary.sort_values('missing_count', ascending=False)
display(missing_summary.head(15))

# ===============================
# 7️ Basic statistics
# ===============================
print("\n" + "="*25, "Basic Statistics (first 20 columns)", "="*25)
display(df.describe(include='all').T.head(20))

# ===============================
# 8️ Additional checks
# ===============================
print("\n" + "="*25, "Additional Quality Checks", "="*25)
num_duplicates = df.duplicated().sum()
print(f"Duplicate Rows: {num_duplicates:,}")
print("\n Column Type Overview:")
print(df.dtypes.value_counts())

memory_mb = df.memory_usage(deep=True).sum() / (1024**2)
print(f"\n Memory Usage: {memory_mb:.2f} MB")

missing_summary.to_csv(r"C:\Users\gowda\Downloads\FDS PROJECT\Airline_Flight_Delay_Prediction\dataset\missing_value_summary.csv", index=False)
print(" Missing value summary saved → missing_value_summary.csv")

# ===============================
# 9️ Save snapshot
# ===============================
print("\n" + "="*25, "Saving Clean Snapshot", "="*25)
df.to_csv(SNAPSHOT_FILE, index=False)
print(f" Snapshot saved → {SNAPSHOT_FILE}")

# ===============================
#  Completion summary
# ===============================
print("\n" + "="*25, "Data Loading Complete", "="*25)
print(f"Rows: {len(df):,}")
print(f"Columns: {df.shape[1]}")
print(f"Missing values: {df.isna().sum().sum():,}")
print(f"Duplicates: {num_duplicates:,}")
print(f"Memory: {memory_mb:.2f} MB")
print("\n Data loading and inspection complete.")
```
